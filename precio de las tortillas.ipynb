```python
# importamos padas ya con el dataset
import pandas as pd

# creamos el CSV
df = pd.read_csv("dataset/tortilla_prices.csv")

# filtramos los precios mayores a 20
df_filtrado = df[df['Price per kilogram'] > 20]

# ordenamos los precios de mayor a menor
df_final = df_filtrado.sort_values('Price per kilogram', ascending=False)

# Imprimir el resultado final
print(df_final)
```
